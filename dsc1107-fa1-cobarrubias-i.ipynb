What relationship does it suggest between carat and cut?

- The diagram displays an inverse correlation between the cut quality and carat.

```r
ggplot(diamonds, aes(x = cut, y = carat, fill = cut)) +
  geom_boxplot() +
  labs(title = "Distribution of Carat Weight by Cut Quality",
       x = "Cut",
       y = "Carat") +
  theme_minimal()
```

What do you conclude? How does this explain the paradoxical trend we found in the plot below?


```r
ggplot(diamonds, aes(x = carat, y = price, color = cut)) +
  geom_point(alpha = 0.3)
  labs(title = "Price vs. Carat Weight by Cut",
       x = "Carat",
       y = "Price") +
  theme_minimal()
```

As shown in this plot, the carat weight is the main driver of the price. It shows an exponential growth in price as the carat weight grows. Due to this, diamond cutters would tend to sacrifice the quality of the cut to preserve the carat weight of the diamond. The boxplot also justifies this conclusion.

```r
ggplot(diamonds, aes(x = carat, y = x)) +
  geom_point(alpha = 0.3)
  labs(title = "Price vs. Carat Weight by Cut",
       x = "Carat",
       y = "Price") +
  theme_minimal()
```

The plot shows that as the carat weight of the diamond grows, the length of the diamond (denoted as x in the given dataset) also grows. The created curve, however, also shows a non-linear growth between the carat weight and the length. This could be attributed to the fact that as the x-length of the diamond grows, its y-width and z-height also grow, meaning the carat weight is correlated to all 3 dimensions. This is consistent with the shape of the plot as it shows a cubic shape of a curve.
