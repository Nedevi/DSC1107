```r
who2 <- who %>% 
    pivot_longer (cols = new_sp_m014:newrel_f65,
    names_to = "key", 
    values_to = "cases", 
    values_drop_na = TRUE)
who2
```

    3.1.3

```r
who3 <- who2 %>% 
  mutate(key = stringr::str_replace(key, "newrel", "new_rel")) %>% 
  separate(key, c("new", "type", "sexage"), sep = "_") %>% 
  select(-new)
who3
```

3.1.4

```r
who_tidy <- who3 %>% 
  separate(sexage, c("sex", "age"), sep = 1)
who_tidy
```

 3.2.1

The variables are country, year, and population

3.2.2 

```r
population3 <- population %>% 
  pivot_longer(
    cols = -c(Country.Name, Country.Code, Indicator.Name, Indicator.Code),
    names_to = "year", 
    values_to = "population",
    names_prefix = "X" 
  )

head(population3)
```

3.2.3

```r
population_tidy <- population3 %>% 
  mutate(year = as.integer(year)) %>% 
  rename(country = Country.Name) %>%
  mutate(country = ifelse(country == "United States", "United States of America", country))

head(population_tidy)
```

```r
# 1. Get the unique country lists
who_countries <- unique(who_tidy$country)
pop_countries <- unique(population_tidy$country)

# 2. Find countries in WHO that are NOT in Population
# (These are the ones you are losing in the join!)
lost_countries <- setdiff(who_countries, pop_countries)

print(lost_countries)
```

3.4.1 - 3.4.3

```r
tuberculosis <- tuberculosis %>% 
    select(-iso2, -iso3) %>% 
  
  filter(!is.na(cases), !is.na(population))
head(tuberculosis)
```

4.1

```r
us_tb_21st_century <- tuberculosis %>% 
  filter(country == "United States of America", year >= 2000) %>% 
  group_by(sex) %>% 
  summarise(total_cases = sum(cases))

print(us_tb_21st_century)
```

4.2

```r
tuberculosis <- tuberculosis %>% 
  mutate(cases_per_100k = (cases / population) * 100000)

head(tuberculosis)
```

4.3

```r
highest_rate <- tuberculosis %>% 
  arrange(desc(cases_per_100k)) %>% 
  slice(1) %>% 
  select(country, year, sex, age, cases_per_100k)

print(highest_rate)

lowest_rate <- tuberculosis %>% 
  arrange(cases_per_100k) %>% 
  slice(1) %>% 
  select(country, year, sex, age, cases_per_100k)

print(lowest_rate)
```

5.1

```r
trends_data <- tuberculosis %>% 
  filter(country %in% c("China", "India", "United States of America")) %>% 
  group_by(country, year, population) %>% 
  summarise(total_cases = sum(cases, na.rm = TRUE), .groups = "drop") %>% 
  mutate(rate = (total_cases / population) * 100000)

ggplot(trends_data, aes(x = year, y = rate, color = country)) +
  geom_line(size = 1.2) +
  scale_y_log10() +
  labs(title = "TB Cases per 100k over Time",
       y = "Cases per 100k (Log Scale)",
       x = "Year") +
  theme_minimal()
```

5.2

```r
age_data <- tuberculosis %>% 
  group_by(country, year, age, population) %>% 
  summarise(cases = sum(cases, na.rm = TRUE), .groups = "drop") %>% 
  mutate(rate = (cases / population) * 100000)

ggplot(age_data, aes(x = age, y = rate)) +
  geom_boxplot(outlier.shape = NA, fill = "lightblue") + # Hide extreme outliers for clarity
  scale_y_log10() +
  labs(title = "Distribution of TB Rates by Age Group",
       x = "Age Group",
       y = "Cases per 100k (Log Scale)") +
  theme_minimal()
```

5.3

```r
data_2000 <- tuberculosis %>% 
  filter(year == 2000) %>% 
  group_by(country, population) %>% 
  summarise(total_cases = sum(cases, na.rm = TRUE), .groups = "drop") %>% 
  mutate(rate = (total_cases / population) * 100000)

ggplot(data_2000, aes(x = population, y = rate)) +
  geom_point(alpha = 0.6) +
  scale_x_log10() + 
  scale_y_log10() + 
  geom_smooth(method = "lm", color = "red", se = FALSE) + 
  labs(title = "TB Rates vs Population Size (Year 2000)",
       x = "Total Population (Log Scale)",
       y = "Cases per 100k (Log Scale)") +
  theme_minimal()
```

Conclusion There is virtually no correlation between a country's population size and its TB rate. Small countries can have high rates (e.g., Djibouti, Eswatini) or low rates (e.g., Iceland). Large countries can also be high (India) or low (USA). The disease burden is driven by healthcare infrastructure and poverty, not simply by how many people live there.
